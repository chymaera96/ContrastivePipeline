```python
import os
import pandas as pd
import numpy as np
```

```python
fpath = "../Desktop/ground_truth/dev.csv"

# Read the csv file into a pandas DataFrame. First row is header.
df = pd.read_csv(fpath, header=0)
df
```

```python
# Get list of unique labels
labels = list(df["labels"].unique())

# Separate each label by commas and get a list of all labels
labels_sep = list(set(",".join(labels).split(",")))
labels_sep
```

```python
label_dict = {'vehicle':[], 'speech': [], 'farm': [], 'home': []}
label_dict['vehicle'] = [
    'Subway_and_metro_and_underground', 'Train', 'Accelerating_and_revving_and_vroom', 'Bus',
    'Vehicle_horn_and_car_horn_and_honking', 'Car_passing_by', 'Motorcycle', 'Car_alarm', 'Car',
    'Engine', 'Traffic_noise_and_roadway_noise', 'Vehicle', 'Motor_vehicle_(road)', 'Truck','Rail_transport']
label_dict['farm'] = ['Livestock_and_farm_animals_and_working_animals', 'Cricket', 'Domestic_animals_and_pets','Chicken_and_rooster','Wind','Dog']
label_dict['speech'] = ['Speech', 'Laughter', 'Crowd', 'Human_voice', 'Conversation', 'Child_speech_and_kid_speaking',
                        'Male_speech_and_man_speaking', 'Female_speech_and_woman_speaking', 'Whispering','Chuckle_and_chortle']
label_dict['home'] = ['Dishes_and_pots_and_pans', 'Sink_(filling_or_washing)', 'Bathtub_(filling_or_washing)',
                      'Domestic_sounds_and_home_sounds','Cupboard_open_or_close','Scissors']
```

```python
# Create a count for each label in label_dict
count_dict = {'vehicle':0, 'speech':0, 'farm':0, 'home':0}
for key in label_dict:
    for label in label_dict[key]:
        count_dict[key] += len(df[df['labels'].str.contains(label)])

count_dict
```

```python
from util import load_config
import json
cfg = load_config('config/baseline.yaml')
test_noise_dir = cfg['test_noise_dir']
# noise_dir = cfg['noise_dir']    # To store index files for each label

# From df get a fname list of all files that contain a label in label_dict
for key in label_dict:
    fname_list = []
    for label in label_dict[key]:
        fname_list += list(df[df['labels'].str.contains(label)]['fname'])
    fname_list = list(set(fname_list))
    # print(fname_list[:10])
    index_list = [os.path.join(test_noise_dir, str(fname)+".wav") for fname in fname_list]
    print(index_list[:10])
    with open(key+".json", 'w') as f:
        json.dump(index_list, f)
```
